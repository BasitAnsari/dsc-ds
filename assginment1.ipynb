```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('real_estate_price_size.csv')
```

```python
data
```

```python
y= data['size']
x1= data['price']
```

```python
data.describe()
```

```python
plt.scatter(x1,y)
plt.xlabel('price')
plt.ylabel('size')
```

```python
x= sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
```

```python
plt.scatter(x1,y)
yhat= 0.0033*x1-122.3349 
fig= plt.plot(x1,yhat,lw=4,c='orange',label='reggLine')
plt.xlabel('price')
plt.ylabel('size')
plt.show()
```
